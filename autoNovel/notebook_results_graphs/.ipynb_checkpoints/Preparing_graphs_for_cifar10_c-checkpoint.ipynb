## List of imports

```python
import wandb
import matplotlib.pyplot as plt
```

## Domain Shift Experiments

```python
api = wandb.Api()# used to import weight and bias pipeline
baseline=api.run('mhaggag96/trends_project/rzfey91s')
cifar10c=api.run('jacopodona/autonovel/3c210uzk')
```

```python
baseline_data=baseline.history()
cifar10c_data=cifar10c.history()
```

```python
fig, ax= plt.subplots(figsize=(15, 10));
ax.plot(baseline_data.cluster_acc_Head_2.tolist(), color='red',label="Baseline")
ax.plot(cifar10c_data.cluster_acc_Head_2.tolist(), color='blue',label="CIFAR10-C")
ax.set_xlabel('Epochs',fontsize=20);
ax.set_ylabel('Cluster accuracy',fontsize=20);
ax.set_title('Head 2 accuracy',fontsize=25);
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(loc='best', framealpha=0.2)
```
